1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 11)
f_arr = 10 * x - 5

h = 1e-5
def f(x):
    return 10 * x - 5 
    
df = (f(x + h) - f(x)) / h

plt.figure(figsize=(10, 5))
plt.plot(x, f_arr, label='f(x) = 10*x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

plt.legend();
```

Кожна додаткова година збільшить оцінку на 10 баллів. Функція зростає рівномірно

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

x = sp.symbols('x')

f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)

# 3. Преобразование в числовые функции
f_sin_func = sp.lambdify(x, f_sin, modules=['numpy'])
sin_deriv_func = sp.lambdify(x, derivative_f_sin, modules=['numpy'])

x_vals = np.linspace(-2*np.pi, 2*np.pi, 500)

y_sin = f_sin_func(x_vals)
y_sin_deriv = sin_deriv_func(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_sin, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals, y_sin_deriv, label="f'(x) = cos(x)", color='red', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()


f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)

f_ln_func = sp.lambdify(x, f_ln, modules=['numpy'])
deriv_ln_func = sp.lambdify(x, derivative_f_ln, modules=['numpy'])

x_vals = np.linspace(0.1, 10, 500) 

y_ln = f_ln_func(x_vals)
y_deriv = deriv_ln_func(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_ln, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals, y_deriv, label="f'(x) = 1/x", color='green', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f = x**2 - 5*x + 1
derivative_f = sp.diff(f, x)

deriv_f = sp.lambdify(x, derivative_f, modules=['numpy'])

x_vals = np.linspace(0, 100, 101) 

y = x_vals**2 - x_vals*5 + 1
y_deriv = deriv_f(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals, y_deriv, label="f'(x) = 1/x", color='green', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
